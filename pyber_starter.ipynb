```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_rides = pd.merge(city_data, ride_data, how="outer", on="city")
# Display the data table for preview
city_rides.head()
#city_data_to_load_df.head()
#ride_data_to_load_df.head()
```

```python
city_rides.dtypes
```

## Bubble Plot of Ride Sharing Data

```python
total_rides = city_rides.groupby("city")["ride_id"].count()
total_rides
```

```python
average_fare = city_rides.groupby("city")["fare"].mean()
average_fare
```

```python
driver_count = city_rides.groupby("city")["driver_count"].mean()
driver_count
```

```python
city_type = city_data.set_index("city")["type"]
city_type.value_counts()
```

```python
pyber_analysis = pd.DataFrame({"Average Fare":average_fare,
                               "Total Number of Rides per City":total_rides,
                              "Number of Drivers": driver_count, 
                              "City Types": city_type})
pyber_analysis.head()
```

```python
urban = pyber_analysis.loc[pyber_analysis["City Types"]=="Urban"]
suburban = pyber_analysis.loc[pyber_analysis["City Types"]=="Suburban"]
rural = pyber_analysis.loc[pyber_analysis["City Types"]=="Rural"]
```

```python
# Obtain the x and y coordinates for each of the three city types
plt.scatter(urban["Total Number of Rides per City"], urban["Average Fare"], color = "lightcoral", edgecolors=
           "black", s = urban["Number of Drivers"]*10, label = "Urban", alpha = 0.5, linewidth = 1.5)
plt.scatter(suburban["Total Number of Rides per City"], suburban["Average Fare"], color = "lightskyblue", edgecolors=
           "black", s = suburban["Number of Drivers"]*10, label = "Suburban", alpha = 0.5, linewidth = 1.5)
plt.scatter(rural["Total Number of Rides per City"], rural["Average Fare"], color = "gold", edgecolors=
           "black", s = rural["Number of Drivers"]*10, label = "Rural",alpha = 0.5, linewidth = 1.5)

# Build the scatter plots for each city types

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rider per city")
plt.ylabel("Average Fare $")
plt.grid()
# Create a legend
plt.legend(loc= "upper right")

# Incorporate a text label regarding circle size
plt.text(45, 30, "Note: Circle size correlates with driver count per city")
# Save Figure
plt.savefig("Pyber Ride Sharing Data")
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_type_pie = city_rides.groupby(["type"])["fare"].sum()
fare_type_pie
# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Fares by City Type")
plt.pie(fare_type_pie, labels= labels, explode = explode, colors = colors, autopct="%1.1f%%", shadow=True, startangle=300)
plt.axis("equal")
# Save Figure and show figure
plt.show()
plt.savefig("% of Total Fares by City Type")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = city_rides.groupby(["type"])["ride_id"].count()
total_rides
# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Drivers by City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.axis("equal")
plt.show()
# Save Figure
plt.savefig("% of Total Drivers by City Type")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = city_data.groupby(["type"])["driver_count"].sum()
# Build Pie Charts
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.title("% of Total Drivers by City Type")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.axis("equal")
# Save Figure
plt.show()
plt.savefig("% of Total Drivers by City Type")
```
